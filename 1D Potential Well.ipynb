```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Constants
hbar = 1.0  
m = 1.0    
L = 1.0   
N = 1000   

dx = L / (N - 1)  
x = np.linspace(0, L, N)  

V = np.zeros(N)
H = np.zeros((N, N))

for i in range(1, N - 1):
    H[i, i] = -2.0 / dx**2 + V[i]
    H[i, i - 1] = 1.0 / dx**2
    H[i, i + 1] = 1.0 / dx**2

#Solving for eigenvalues
energies, wavefunctions = np.linalg.eigh(H)

wavefunctions = wavefunctions / np.sqrt(dx)


n_levels = 4

plt.figure(figsize=(10, 6))
for n in range(n_levels):
    plt.plot(x, wavefunctions[:, n]**2, label=f"n={n+1}, E={energies[n]:.3f}")

plt.title("Probability Densities of 1D Infinite Potential Well")
plt.xlabel("Position x")
plt.ylabel("Probability Density |Psi(x)|^2")
plt.legend()
plt.grid()
plt.show()


print("Energy eigenvalues (in natural units):")
for n in range(n_levels):
    print(f"n={n+1}: E={energies[n]:.3f}")
```
